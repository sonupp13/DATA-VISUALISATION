**2. Basic Plotting with Matplotlib**


```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Line Plot')
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

#the parameter " s "  represents the size of the markers used in the scatter plot
plt.scatter(x, y, marker='x', color='blue', s=100)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')
plt.show()
```

```python
import matplotlib.pyplot as plt

categories = ['Category A', 'Category B', 'Category C']
values = [10, 20, 15]

plt.bar(categories, values, color='green')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.randn(1000)  # Generating random data
plt.hist(data,color='orange' ,bins=5)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Predefined data
data = [1 , 2 , 2, 3,  3,  3,  4 ,  4,  5,  6  ,6 , 6]
# Creating a histogram
plt.hist(data,color='skyblue', edgecolor='black' , bins=6 ) #default bin = 10
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram of Predefined Data')
plt.show()
```

```python
import pandas as pd

import matplotlib.pyplot as plt

df1 = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/titanic.csv")
```

```python
df1.head(3)
```

```python
df1['Sex'].value_counts()
```

```python
# Counting passengers in each class
class_counts = df1['Sex'].value_counts()

# Creating a bar chart
plt.bar(class_counts.index, class_counts.values)
plt.xlabel('Passenger Class')
plt.ylabel('Number of Passengers')
plt.title('Passenger Count by Class')
plt.show()
```

```python
# Removing missing values in the 'Age' column for plotting
age_data = df1['Age'].dropna()
print(age_data)


# Creating a histogram for age distribution
plt.hist(age_data, bins=20, color='orange', edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Passenger Age Distribution')
plt.show()
```

```python
# Removing missing values in the 'Fare' column for plotting
valid_data = df1[['Age', 'Fare']].dropna()

age_data = valid_data['Age']
fare_data = valid_data['Fare']

plt.scatter(age_data, fare_data, color='green', marker='.')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Age vs Fare Paid')
plt.show()
```
