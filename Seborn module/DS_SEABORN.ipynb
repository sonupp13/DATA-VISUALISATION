```python
import seaborn as sns
df=sns.load_dataset("tips")
df
```

```python
df.head()
```

**Correlation with Heatmap**
A correlation heatmap uses colored cells, typically in a monochromatic scale, to show a 2D correlation matrix (table) between two discrete dimensions or event types. It is very important in Feature Selection. (interralation between features)

```python
d1=df.corr()
d1
```

Correlation typically refers to the statistical relationship between two variables. It indicates how much and in what way one variable changes with respect to another. Correlation measures both the strength and direction of the linear relationship between two variables.

1: Perfect positive relationship—variables increase together.

0: No relationship—changes in one don't link to changes in the other.

-1: Perfect negative relationship—one variable increases, the other decreases proportionally.

```python
sns.heatmap(d1,annot= True)
```

**JoinPlot**
A join plot allows to study the relationship between 2 numeric variables. The central chart display their correlation. It is usually a scatterplot, a hexbin plot, a 2D histogram or a 2D density plot

```python
sns.jointplot(x="tip",y="total_bill", data=df,kind="hex")
```

```python
sns.jointplot(x="tip",y="total_bill", data=df,kind="reg",color="r")
```

**Pair plot**
A “pairs plot” is also known as a scatterplot, in which one variable in the same data row is matched with another variable's value, like this: Pairs plots are just elaborations on this, showing all variables paired with all the other **variables**

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue="sex")
```

**Dist plot**
Dist plot helps us to check the distribution of the columns feature

```python
sns.distplot(df["tip"])
```

```python
sns.distplot(df['tip'],kde=False,bins=10)
```

Categorical Plots
Seaborn also helps us in doing the analysis on Categorical Data points. In this section we will discuss about

boxplot
violinplot
countplot
bar plot

```python
import seaborn as sns
df=sns.load_dataset("tips")
df
```

```python
sns.countplot(df)
```

```python
sns.countplot(x='sex',data=df)
```

```python
sns.countplot(y='sex',data=df)
```

```python
# barplot
sns.barplot(x='sex',y='total_bill',data=df)
```

```python
df.head()
```

**Box plot**
A box and whisker plot (sometimes called a boxplot) is a graph that presents information from a five-number summary.

```python
sns.boxplot(x='smoker',y='total_bill',data=df)
```

```python
sns.boxplot(x='day',y='total_bill',data=df,palette='rainbow')
```

```python
sns.boxplot(data=df)
```

```python
sns.boxplot(x='total_bill',y='day',data=df,hue='smoker')
```

```python
Violin Plot
Violin plot helps us to see both the distribution of data in terms of Kernel density estimation and the box plot
```

```python
sns.violinplot(x='day',y='total_bill',data=df,palette='rainbow')
```
